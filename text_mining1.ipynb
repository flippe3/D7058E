Loop to remove duplicates from list (Pandas library has a special function for this when dealing with data)

```python
a = [10,20,30,20,10,50,60,40,80,50,40]
dup_items = set() 
uniq_items = [] 

for x in a:                         # List comprehension is a more efficient way to do something similar
    if x not in dup_items:
        uniq_items.append(x)
        dup_items.add(x) 

print(dup_items)
print(uniq_items)
```

Functions

```python
# Get the second smallest number in a list

def second_smallest(numbers): 
    a1, a2 = float('inf'), float('inf') 
    for x in numbers: 
        if x <= a1: 
            a1, a2 = x, a1 
        elif x < a2: 
            a2 = x 
    return a2 

res = second_smallest([-1, 4, -3, 1, 5, 2, -8, -2, 0])
print(res)
```

Text Preprocessing with the Pandas library

```python
import pandas as pd

def preprocess_pandas(data):
    data['Sentence'] = data['Sentence'].str.lower()                                                                     # make all lowercase
    data['Sentence'] = data['Sentence'].replace('[a-zA-Z0-9-_.]+@[a-zA-Z0-9-_.]+', '', regex=True)                      # remove emails
    data['Sentence'] = data['Sentence'].replace('((25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)(\.|$)){4}', '', regex=True)    # remove IP address
    data['Sentence'] = data['Sentence'].str.replace('[^\w\s]','')                                                       # remove special characters
    data['Sentence'] = data['Sentence'].replace('\d', '', regex=True)                                                   # remove numbers
    return data

lst = ['Det är mycket bra att laga mat','Swenglish is fun to speak! :)']
df = pd.DataFrame (lst, columns = ['Sentence'])
data = preprocess_pandas(df)
print(data)
```

## Task 1.3 IMDB visualization with t-SNE 

```python
from nltk.corpus import brown
from gensim.models import Word2Vec
df = pd.read_csv("IMDB Dataset.csv")[:20]

word_vectors = []
y = []
y2 = []
counter = 0

model = Word2Vec(brown.sents())

for i in df['review']:
    tokens = word_tokenize(i)
    tags = pos_tag(tokens, tagset='universal')
    for words in tags:
        try: 
            word_vectors.append(model.wv[words[0]])
            y2.append(words[1])
            y.append(int(df['sentiment'][counter]=='positive'))
        except:
            continue
    counter += 1
    
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import numpy as np

X = np.array(word_vectors)

print(X[0].shape)

tsne = TSNE(n_components=2)
Z = tsne.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

new_df = pd.DataFrame()
new_df['y'] = y2 # y is just the sentiment classes
new_df['x_1'] = Z[:,0]
new_df['x_2'] = Z[:,1]

sns.scatterplot(x="x_1", y="x_2", hue='y', palette=sns.color_palette("hls", 12), data=new_df).set(title="TSNE Visualization word2vec embeddings") 
```
